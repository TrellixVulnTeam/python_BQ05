```python
import os
import os.path as path
from tqdm import tqdm

import funclib.iolib as iolib
import funclib.baselib as baselib

def splitf(s):
    return os.path.dirname(s), os.path.splitext(os.path.basename(s))[0], os.path.splitext(s)[1]
```

```python
#cleans spectrum roms prioritising by EXTENSIONS order
EXTENSIONS = {'.d64':100, '.t64':80, '.tap':60, '.prg':50, '.p00':40}

root = 'C:/temp/c64_best_games'
#root = 'C:/temp/test'


def get_score(fname):
    d, f, e = iolib.get_file_parts2(fname)
    score = EXTENSIONS.get(e.lower(), 0) - len(f)
    return score
    
    

for f in os.listdir(root): #game folders, 
    D = {}
    fld_game = os.path.normpath(os.path.join(root, f)) #fld_game e.g. C:/temp/c64_best_games/paradroid
    if not os.path.isdir(fld_game): continue

    for fname in os.listdir(fld_game):
        fname = os.path.normpath(os.path.join(fld_game, fname))
        if not os.path.isfile(fname): continue
        d, f, e = splitf(fname)
        e = e.lower()
        if not e in EXTENSIONS: continue  
        score = get_score(fname)
        D[fname] = int(score)  # {'c:/temp/games/paradroid.d64':90, 'c:/temp/games/paradroid.t64':70, ...}
    
    #print(D)

    for k, v in D.items():
        if v < max(D.values()):
            try:
                os.remove(k)
                print('deleted %s ' % k)
            except:
                pass
            
```
