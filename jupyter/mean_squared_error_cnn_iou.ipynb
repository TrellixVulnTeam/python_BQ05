```python
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
from dblib import mssql
```

```python
#NAS
sql = "" \
"select sample_lengthid ,cnn as CNN ,status" \
",case when transform='hflip' then 'horizontal flip' else 'none' end as transform ,accuracy as Score ,IoU" \
" from  v_mv_long_form" \
" where not accuracy is null and status='Success' and (transform='None' or transform='hflip') and cnn='nas'"


with mssql.Conn('imagedb', '(local)') as cnn:
    df_nas = pd.read_sql(sql, cnn)
list(df_nas)
nas_iou = np.array(df_nas.IoU)
nas_pred = np.ones(nas_iou.shape[0])
mse_nas = mean_squared_error(nas_pred, nas_iou)
```

```python
#RES
sql = "" \
"select sample_lengthid ,cnn as CNN ,status" \
",case when transform='hflip' then 'horizontal flip' else 'none' end as transform ,accuracy as Score ,IoU" \
" from  v_mv_long_form" \
" where not accuracy is null and status='Success' and (transform='None' or transform='hflip') and cnn='res'"


with mssql.Conn('imagedb', '(local)') as cnn:
    df_res = pd.read_sql(sql, cnn)
list(df_res)
res_iou = np.array(df_res.IoU)
res_pred = np.ones(res_iou.shape[0])
mse_res = mean_squared_error(res_pred, res_iou)
```

```python
#SSD
sql = "" \
"select sample_lengthid ,cnn as CNN ,status" \
",case when transform='hflip' then 'horizontal flip' else 'none' end as transform ,accuracy as Score ,IoU" \
" from  v_mv_long_form" \
" where not accuracy is null and status='Success' and (transform='None' or transform='hflip') and cnn='ssd'"

with mssql.Conn('imagedb', '(local)') as cnn:
    df_ssd = pd.read_sql(sql, cnn)
list(df_ssd)
ssd_iou = np.array(df_ssd.IoU)
ssd_pred = np.ones(ssd_iou.shape[0])
mse_ssd = mean_squared_error(ssd_pred, ssd_iou)
```

```python
print('nas:%.3f   res:%.3f   ssd:%.3f   ' % (mse_nas, mse_res, mse_ssd))
```

```python
#Calc mean absolute error between flp and not flip
#SSD
sql = "" \
"select sample_lengthid ,cnn as CNN ,status" \
",case when transform='hflip' then 'horizontal flip' else 'none' end as transform ,accuracy as Score ,IoU" \
" from  v_mv_long_form" \
" where not accuracy is null and status='Success' and transform='None'"

with mssql.Conn('imagedb', '(local)') as cnn:
    df_none = pd.read_sql(sql, cnn)

none_iou = np.array(df_none.IoU)
none_pred = np.ones(none_iou.shape[0])
```

```python
#Calc mean absolute error between flp and not flip
#SSD
sql = "" \
"select sample_lengthid ,cnn as CNN ,status" \
",case when transform='hflip' then 'horizontal flip' else 'none' end as transform ,accuracy as Score ,IoU" \
" from  v_mv_long_form" \
" where not accuracy is null and status='Success' and transform='hflip'"

with mssql.Conn('imagedb', '(local)') as cnn:
    df_flip = pd.read_sql(sql, cnn)

flip_iou = np.array(df_flip.IoU)
flip_pred = np.ones(flip_iou.shape[0])
```

```python
s = 'flip mae:%0.3f   untransformed mae:%0.3f' % (np.mean(flip_pred-flip_iou),
                              np.mean(none_pred-none_iou))
print(s)
```
