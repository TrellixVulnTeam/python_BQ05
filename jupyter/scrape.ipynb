```python
from bs4 import BeautifulSoup as BS
import requests
print('Done')
```

```python
url='http://www.seagullfishingtackle.com/p/beach-catch-reports.html'
```

```python
soup =  BS(requests.get(url, timeout=60).content, 'html.parser')
print('Done')
```

```python
soup_titles = soup.find_all(style=['color: blue; font-family: "arial" , "sans-serif"; font-size: 24.0pt;', "color: blue; font-size: xx-large;", "color: blue; font-size: x-large;"])
titles = [s.get_text() for s in soup_titles]
```
